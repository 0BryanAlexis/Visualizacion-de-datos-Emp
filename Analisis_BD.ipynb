```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Ignorar advertencias
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/VictorMM7/Optativa-Profesional-Data-Science/refs/heads/main/student_performance.csv', sep = ',' )

dataset.head()
```

```python
dataset.dtypes
```

```python
print("Estadisticas descriptivas basicas")
print(dataset.describe())
```

```python
numeric_columns = ['AttendanceRate', 'StudyHoursPerWeek', 'PreviousGrade', 'ExtracurricularActivities', 'FinalGrade']

print("\nMedidas de tendencia central para columnas numericas especificas:")
for  column in numeric_columns:
    print(f"\n{column}:")
    print(f"Media: {dataset[column].mean()}")
    print(f"Mediana: {dataset[column].median()}")
    print(f"Moda: {dataset[column].mode()[0]}")
    print(f"Desviacion estandar: {dataset[column].std()}")
    print(f"Minimo: {dataset[column].min()}")
    print(f"Maximo: {dataset[column].max()}")
    print(f"Cuartiles: \n{dataset[column].quantile([0.25, 0.5, 0.75])}")
```

```python
#Grafico de dispersion
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(10,4))
plt.scatter(x, y)
plt.title('Grafico de Dispersion')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Se selecciona solo las columnas numéricas, de otra forma, genera error
numeric_columns = dataset.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap de correlación')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x="Gender", y="AttendanceRate", data=dataset)
plt.title('Violin plot: Calificaciones den Ultimo Grado')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="Gender", y="FinalGrade", hue="Gender", data=dataset)
plt.title('Box plot: Total de la cuenta por día y si es fumador')
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.pairplot(dataset, hue="Gender")
plt.suptitle('Pair plot de variables en el conjunto de datos tips', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x="StudyHoursPerWeek", y="FinalGrade", data=dataset)
plt.title('Regresión lineal: Propina vs Total de la cuenta')
plt.show()
```
